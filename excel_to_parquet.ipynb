```python
import pandas as pd
import pyarrow
import os

# --- RUTAS DE ARCHIVOS ---
ruta_csv = r"C:\Users\facuc\OneDrive\Escritorio\GCBA\csv\Formulario kobo hoja 1.csv"
ruta_directorio_parquet = r"C:\Users\facuc\OneDrive\Escritorio\GCBA\Parquet"
nombre_archivo_parquet = "Formulario_KOBO_v4.parquet"
ruta_completa_parquet = os.path.join(ruta_directorio_parquet, nombre_archivo_parquet)

# -----------------------------------------------------
# --- PASO 1: CARGAR EL ARCHIVO CSV ---
# -----------------------------------------------------
df_hist = None  # inicializamos

try:
    df_hist = pd.read_csv(
        ruta_csv,
        dtype=str,              # todo como string para evitar problemas
        on_bad_lines='skip',    # ignora filas con n√∫mero de columnas incorrecto
        sep=',',                # separador (cambia si tu CSV usa otro)
        encoding='utf-8'        # codificaci√≥n
    )
    print(f"‚úÖ Archivo CSV cargado exitosamente desde: {ruta_csv}")
    print(f"üìä Total filas cargadas: {len(df_hist)}")
except FileNotFoundError:
    print(f"‚ùå Error: No se encontr√≥ el archivo CSV en la ruta: {ruta_csv}")
    exit()
except Exception as e:
    print(f"‚ùå Ocurri√≥ un error al leer el archivo CSV: {e}")
    exit()

# -----------------------------------------------------
# --- PASO 2: LIMPIEZA Y PREPARACI√ìN ---
# -----------------------------------------------------
if df_hist is not None:
    columnas_object = df_hist.select_dtypes(include=["object"]).columns
    print("üõ†Ô∏è Convirtiendo columnas 'object' a string...")
    for col in columnas_object:
        df_hist[col] = df_hist[col].astype(str)
    print("üõ†Ô∏è Conversi√≥n de tipos completada.")
else:
    print("‚ùå Error: df_hist no se pudo crear.")
    exit()

# -----------------------------------------------------
# --- PASO 3: GUARDAR A PARQUET ---
# -----------------------------------------------------
os.makedirs(ruta_directorio_parquet, exist_ok=True)

try:
    df_hist.to_parquet(
        ruta_completa_parquet,
        engine="pyarrow",
        compression="snappy",
        index=False
    )
    print(f"‚úÖ El DataFrame se guard√≥ exitosamente como Parquet en: {ruta_completa_parquet}")
except Exception as e:
    print(f"‚ùå Ocurri√≥ un error al intentar guardar el archivo Parquet: {e}")


```
